**Exerc√≠cio 1.** Fa√ßa o gr√°fico da fun√ß√£o
$ùëì(ùë•)=sin‚Å°(ùë•^2)+cos‚Å°(x^2)$
entre $‚àí6$ (incluso) e $6$ (n√£o incluso) com incrementos de 0.1.
Calcule os valores de $ùëì(ùë•)$ para essa faixa.
Plote o gr√°fico da fun√ß√£o.

```python
import numpy as np
import matplotlib.pyplot as plt

# Criar a faixa de valores de x
x = np.arange(-6, 6, 0.1)

# Calcular os valores de f(x)
y = np.sin(x**2) + np.cos(x)**2

# Plotar a fun√ß√£o
plt.plot(x, y, label='$f(x) = \sin(x^2) + \cos^2(x)$')
plt.title('Gr√°fico de $f(x) = \sin(x^2) + \cos^2(x)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

**Exerc√≠cio 2.** Fa√ßa o gr√°fico da fun√ß√£o $f(x)=2x^3‚àí4x^2+x‚àí5$ para $x$ entre $-3$ e $3$ com incrementos de $0.1$. Calcule os valores de $f(x)$ para essa faixa.
Plote o gr√°fico da fun√ß√£o.

```python
import numpy as np
import matplotlib.pyplot as plt

# Criar a faixa de valores de x
x = np.arange(-3, 3, 0.1)

# Calcular os valores de f(x)
y = 2*x**3 - 4*x**2 + x - 5

# Plotar a fun√ß√£o
plt.plot(x, y, label='$f(x) = 2*x**3 - 4*x**2 + x - 5 ')
plt.title('Gr√°fico de $f(x) = 2*x**3 - 4*x**2 + x - 5 ')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```


**Exerc√≠cio 3:** Fa√ßa o gr√°fico da fun√ß√£o $ùëì(ùë•)=ùëéùë•^2+ùëèùë•+ùëê$ com valores e intervalo definidos pelo usu√°rio.
1. Inicialmente, pe√ßa ao usu√°rio para fornecer os valores de $a,b,c,$ o in√≠cio do intervalo, $r1$, e o fim do intervalo, $r2$.
2. Crie uma faixa de valores de $x$ com incrementos de $0.1$ no intervalo fornecido.
3. Calcule os valores de $ùëì(ùë•)$ para essa faixa.
4. Plote o gr√°fico da fun√ß√£o.

```python
import numpy as np
import matplotlib.pyplot as plt
#entradas
a =float(input('Digite o valor de a:'))
b=float(input('Digite o valor de b:'))
c=float(input('Digite o valor de c:'))
r1=float(input('Digite o valor de r1:'))
r2=float(input('Digite o valor de r2:'))

# Criar a faixa de valores de x
x = np.arange( r1,r2,  0.1)

# Calcular os valores de f(x)
y = a*x**2 - b*x + c

# Plotar a fun√ß√£o
plt.plot(x, y, label='$f(x)= a*x**2 - b*x + c')
plt.title('Gr√°fico de $f(x) = a*x**2 - b*x + c ')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

 **Exerc√≠cio 4.** Fa√ßa o gr√°fico das fun√ß√µes $f_1(x) = sin(ax)$ e $f_2(x) = cos(bx)$ em um intervalo definido pelo usu√°rio.

1. Inicialmente, pe√ßa ao usu√°rio para fornecer os valores de $a,b$  e os valores de in√≠cio do intervalo, $r1$, e o fim do intervalo, $r2$.
2. Crie uma faixa de valores de $x$ com incrementos de $0.1$ no intervalo fornecido.
3. Calcule os valores de $ùëì_1(ùë•)$ e $f_2(x)$ para essa faixa.

```python
#entrada
a= float(input('Forne√ßa o valor de a: '))
b=float(input('Forne√ßa o valor de b: '))
r1=float(input('Forne√ßa o valor de r1: '))
r2=float(input('Forne√ßa o valor de r2: '))
#valorx
x = np.arange(r1,r2,0.1)
#calculo
y1=np.sin(a*x)
y2=np.cos(b*x)
#grafico
plt.plot(x,y1,label='$f1(x)= sin(a*x)')
plt.plot(x,y2,label='$f1(x)= cos(b*x)')
plt.title('Grafico f1 e f2')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()


```
