exercicio 7- Dado o conjunto de dados do Exercício 7, onde são resumidos os dados de venda de uma loja e cada linha representa as vendas de um produto em um determinado dia, adicione as seguintes colunas:

Receita Total: o produto do número de unidades vendidas pelo preço unitário.

Desconto Aplicado: calcule um desconto de 10% sobre a Receita Total e adicione essa nova coluna ao DataFrame.

Receita Líquida: a Receita Total menos o Desconto Aplicado.

```python
#7
import pandas as pd
df = pd.read_csv(" https://drive.google.com/uc?id=1-VL_ZQLFCt9Igip-CHbqn5AE2y1g1OsD&export=download")
df['Receita Total'] = df['Preço Unitário']*df['Unidades Vendidas']
df['Desconto Aplicado']=df['Receita Total']*0.1
df['Receita Líquida']=df['Receita Total']-df['Desconto Aplicado']
print(df)
```
